```python
sensor_reading=[45,78,-12,90,105,66,88]
def clean_data(data):
  cleaned=[]
  for value in data:
      if 0 <= value <= 100:
          cleaned.append(value)
  return cleaned
def calculate_average(data):
    if len(data)==0:
      return 0
    return sum(data)/len(data)
cleaned_readings=clean_data(sensor_reading)
average_value=calculate_average(cleaned_readings)

print("cleaned_data:",cleaned_readings)
print("average:",average_value)
```

```python
students={
    "ali:":[65,70,80],
    "sara":[90,85,88],
    "ahmed":[40,55,60]
}
def calculate_average(marks):
  if len(marks)==0:
    return 0
    return sum(marks)/len(marks)

def has_passed(average):
  return average>=50
def display_summary(students_disct):
  for name,marks in students_disct.items():
    avg=sum(marks)/len(marks)
    if has_passed(avg):
      status="pass"
    else:
      status="fail"
    print("student:",name)
    print("marks:",marks)
    print("average:",round(avg,2))
    print("result",status)

display_summary(students)
```

```python
class RuleBasedClassifier:
    def __init__(self, threshold):
        """
        Initialize the classifier with a threshold value.
        Values >= threshold → True (positive class)
        Values < threshold  → False (negative class)
        """
        self.threshold = threshold

    def classify_single(self, value):
        """
        Classify a single value.
        Returns True if value >= threshold, False otherwise.
        """
        return value >= self.threshold

    def classify_list(self, values):
        """
        Classify a list of values.
        Returns a list of booleans with the same length as input.
        """
        return [self.classify_single(x) for x in values]

threshold = 60
values = [45, 72, 88, 30, 65]
classifier = RuleBasedClassifier(threshold)
print("Single classifications:")
print(45, "→", classifier.classify_single(45))
print(72, "→", classifier.classify_single(72))
print("\nList classification:")
result = classifier.classify_list(values)
print("Values :", values)
print("Classes:", result)
print("\nDetailed view:")
for v, c in zip(values, result):
    print(f"{v:3} → {c}")
```

```python
def clean_numeric_data(raw_values, min_val=None, max_val=None, remove_negatives=False):
    cleaned = []

    for value in raw_values:

        if not isinstance(value, (int, float)):
            continue

        if remove_negatives and value < 0:
            continue
        if min_val is not None and value < min_val:
            continue
        if max_val is not None and value > max_val:
            continue

        cleaned.append(value)

    return cleaned


class CleanedDataset:

    def __init__(self, cleaned_values):
        if not all(isinstance(x, (int, float)) for x in cleaned_values):
            raise ValueError("CleanedDataset can only store numbers")
        self.values = cleaned_values
        self._n = len(cleaned_values)

    def count(self):
        return self._n

    def min(self):
        return min(self.values) if self.values else None

    def max(self):
        return max(self.values) if self.values else None

    def mean(self):
        if self._n == 0:
            return None
        return sum(self.values) / self._n

    def median(self):
        if self._n == 0:
            return None
        sorted_vals = sorted(self.values)
        mid = self._n // 2
        if self._n % 2 == 1:
            return sorted_vals[mid]
        else:
            return (sorted_vals[mid-1] + sorted_vals[mid]) / 2

    def summary(self):
        return {
            "count": self.count(),
            "min": self.min(),
            "max": self.max(),
            "mean": self.mean(),
            "median": self.median(),
            "values": self.values
        }

raw_data = [10, 55, -3, 78, 120, 66, 92, "invalid", 999]
cleaned = clean_numeric_data(
    raw_data,
    min_val=0,
    max_val=200,
    remove_negatives=True
)

dataset = CleanedDataset(cleaned)
print("Cleaned values:", dataset.values)
print("\nSummary statistics:")
stats = dataset.summary()

for key, value in stats.items():
    if key != "values":
        print(f"{key:>8} : {value}")
    else:
        print(f"{key:>8} : {value}")
print("\nIndividual calls:")
print(f"Mean   = {dataset.mean():.2f}")
print(f"Median = {dataset.median():.2f}")
```
