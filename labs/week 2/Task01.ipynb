```python
monthly_income = [65000, 76000, 78000, 86000]
employment_status = ["employed", "unemployed", "employed", "unemployed"]

eligibility = 0

for i in range(len(monthly_income)):
    if monthly_income[i] >= 60000 and employment_status[i] == "employed":
        eligibility += 1

print("eligibility:", eligibility)
```

#task 5

```python
total_vehicles = int(input("Enter total number of vehicles: "))

violation_count = 0

for i in range(total_vehicles):
    speed = int(input(f"Enter speed of vehicle {i+1}: "))

    if speed > 80:
        print("Overspeeding detected!")
        violation_count += 1

percentage = (violation_count / total_vehicles) * 100
print("Total vehicles:", total_vehicles)
print("Overspeeding vehicles:", violation_count)
print("Violation percentage:", percentage, "%")
```
